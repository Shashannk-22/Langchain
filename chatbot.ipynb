```python
import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv('GROQ_API_KEY')
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.1-8b-instant", api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content="Hey! My name is Shashannk and I am an AI engineer")])
```

```python
from langchain_core.messages import AIMessage
model.invoke([HumanMessage(content="Hey! My name is Shashannk and I am an AI engineer"),
AIMessage(content="Nice to meet you, Shashannk. It's great to know that you're an AI engineer. AI is a rapidly evolving field with a wide range of applications, and I'm sure you're working on some exciting projects. What areas of AI are you particularly interested in or have experience with? Are you into natural language processing, computer vision, or perhaps reinforcement learning?"),
HumanMessage(content="Hey Whats my name and what do I Do?")
])
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store={}

def get_session_id(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]= ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model, get_session_id)
```

```python
config= {"configurable":{"session_id":"chat1"}}
```

```python
response=with_message_history.invoke(
    [HumanMessage(content=" Hi, My name is dhruv and I am an AI engineer")], config=config
)
response.content
```

```python
with_message_history.invoke(
    [HumanMessage(content="Can you remind me whats my name and what do I do?")], config=config
)
```

```python
config1= {"configurable":{"session_id":"chat2"}}
with_message_history.invoke(
    [HumanMessage(content="Can you remind me whats my name and what do I do?")], config=config1
)
```

```python
response=with_message_history.invoke(
    [HumanMessage(content=" Hi, My name is John")], config=config1
)
response.content
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
prompt= ChatPromptTemplate.from_messages(
    [
        "system",
        "You are a helpful assistant answer all questions to the best of your ability",
        MessagesPlaceholder(variable_name="messages"),
    ]
)

chain= prompt | model
```

```python
chain.invoke({"messages":[HumanMessage(content="Hello! My name is Shashannk.")]})
```

```python
with_message_history= RunnableWithMessageHistory(chain, get_session_id)
```

```python
config3= {"configurable":{"session_id":"chat3"}}
response= with_message_history.invoke(
    [HumanMessage(content="Whats my name?")],
    config=config3

)
response.content
```

```python
prompt= ChatPromptTemplate.from_messages(
    [
        "system",
        "You are a helpful assistant answer all questions to the best of your ability in {language}",
        MessagesPlaceholder(variable_name="messages"),
    ]
)

chain= prompt | model
```

```python
response=chain.invoke({"messages":[HumanMessage(content="Hi my name is Shashannk")],"language":"Spanish"})
response.content
```

```python
config4= {"configurable": {"session_id": "chat4"}}
```

```python
with_message_history= RunnableWithMessageHistory(chain, get_session_id, input_messages_key="messages")
response= with_message_history.invoke(
    {'messages': [HumanMessage(content="Hi I am Shashannk")],"language":"Spanish"},
    config=config4
)
response.content
```

```python
from langchain_core.messages import SystemMessage, trim_messages
trimmer=trim_messages(
    max_tokens=70,
    strategy="last",
    token_counter=model,
    include_system=True,
    allow_partial=False,
    start_on="human"
)

messages = [
    SystemMessage(content="you're a good assistant"),
    HumanMessage(content="hi! I'm bob"),
    AIMessage(content="hi!"),
    HumanMessage(content="I like vanilla ice cream"),
    AIMessage(content="nice"),
    HumanMessage(content="whats 2 + 2"),
    AIMessage(content="4"),
    HumanMessage(content="thanks"),
    AIMessage(content="no problem!"),
    HumanMessage(content="having fun?"),
    AIMessage(content="yes!"),
]
trimmer.invoke(messages)
```

```python
from operator import itemgetter
from langchain_core.runnables import RunnablePassthrough

chain=(
    RunnablePassthrough.assign(messages=itemgetter("messages")| trimmer)
    | prompt
    | model
)

chain.invoke({"messages":messages + [HumanMessage(content="What ice cream do I like?")],
"language":"English"})
```

```python
chain.invoke({"messages":messages + [HumanMessage(content="What math problem did I ask about?")],
"language":"English"})
```

```python
with_message_history= RunnableWithMessageHistory(chain, get_session_id, input_messages_key="messages")
config5= {"configurable":{"session_id":"chat5"}}
```

```python
response= with_message_history.invoke(
    {"messages": [HumanMessage(content="what ice cream do I like and what math problem did I ask about?")],
    "language":"English"},
    config=config5
)
response.content
```
